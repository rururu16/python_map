```python
import folium
```

#### folium.Map(location=[緯度, 経度], zoom_start=整数)
指定位置の地図をとってきて表示する（いきなり）

```python
m = folium.Map(location=[35, 135], zoom_start=100)
```

#### folium.Marker([緯度, 経度], popup=文字列, tooltip=文字列)
指定位置にマーカーを配置する。カーソルを重ね合わせる/クリックしたときの表示を指定できる

```python
marker = folium.Marker([35 ,135], popup="わーーー", tooltip="クリック")
marker.add_to(m)
m
```

#### folium.Icon(icon=文字列, color=文字列)
マーカーに設定できるアイコン。iconで形状、colorで色を指定する

```python
icon = folium.Icon(icon="info-sign", color="green")
marker = folium.Marker([35 ,135], popup="わーーー", tooltip="やっほ",icon=icon)
marker.add_to(m)
m
```

#### folium.Map.save(filename)
地図をhtmlに保存

```python
m.save("data/map.html")
```
